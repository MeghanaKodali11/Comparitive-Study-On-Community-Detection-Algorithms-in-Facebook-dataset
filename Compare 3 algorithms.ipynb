```python
import matplotlib.pyplot as plt 
  
# x-coordinates of left sides of bars  

  
# heights of bars 
values = [144.304, 125.115, 178.315]
  
# labels for bars 
names = ['Label Propogation','Fast Unfolding', 'Spectrul Clustering'] 

plt.figure(figsize=(10,10))
# plotting a bar chart 
plt.bar(names, values, width = 0.7) 
  
# naming the x-axis & Y-axis 
plt.xlabel('Community detection algorithms')
plt.ylabel('Time taken in sec to execute the algorithm')
# plot title 
plt.title('Comparision of 3 algorithms based on time taken') 

# function to show the plot 
plt.show() 
```
